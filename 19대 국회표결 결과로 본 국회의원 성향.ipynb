찬성(1), 반대(-1), 미투표(0) 비중을 보면 찬성이 압도적으로 많고 그 다음이 미투표 입니다. 반대표를 던지는 경우는 소수인 걸로 보아 국회에서 이뤄지는 표결이 정당의 당론으로 이뤄지거나 미리 다 약속된 채 이뤄지고 있음을 알게 해줍니다. 그렇다면 의원들의 표결 성향은 정말 당성 또는 여/야로 갈리게 될까요.

```python
df_congress_votes.iloc[:,3:].unstack().reset_index(drop=True).value_counts()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=1).fit(df_congress_votes.iloc[:,3:])
```

각 row 에 트레이닝된 클러스터 번호를 labeling 합니다.

```python
df_congress_votes['labels'] = pd.Series(kmeans.labels_)
```

각 클러스터에 군집한 정당별 국회의원 수를 테이블로 표시합니다. 여당도 제1야당인 더불어민주당도 0번 클러스터에 대부분의 국회의원들이 들어가 있습니다. 정의당 5명 국회의원 모두가 0번 클러스터에 들어있는 걸 보면 의원들의 의결 성향은 야당과 여당으로 갈리는 게 아님을 보여줍니다.

```python
pd.crosstab(df_congress_votes['labels'], df_congress_votes['party'], rownames=['cluster-#'], colnames=['정당명'])
```

## 클러스터 Plot

```python
from sklearn.decomposition import PCA
pca = PCA(2)
```

```python
datapoints = pca.fit_transform(df_congress_votes.iloc[:,3:-1])
centroids = pca.transform(kmeans.cluster_centers_)
```

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
matplotlib.rcParams['figure.figsize'] = (15.0, 8.0)
matplotlib.rcParams['font.family'] = 'NanumGothic'
```

```python
plt.figure(figsize=[30,20])
plt.gca().axes.get_xaxis().set_visible(False)
plt.gca().axes.get_yaxis().set_visible(False)

color_blue = '#4EACC5'
color_red = '#FF9C34' 
colors = map(lambda l:color_blue if l==0 else color_red, df_congress_votes['labels'].values)
plt.scatter(x=datapoints[:,0], y=datapoints[:,1], s=100,c=colors)
plt.scatter(x=centroids[:,0], y=centroids[:,1], s=400, marker='*', c=[color_blue, color_red])

blue_handle = mpatches.Patch(color=color_blue, label='cluster-0')
red_handle = mpatches.Patch(color=color_red, label='cluster-1')
plt.legend(handles=[blue_handle, red_handle])
plt.show()
```

## 국회의원 이름 및 소속 정당 표시

```python
df_sae = df_congress_votes[df_congress_votes.party == '새누리당']
df_min = df_congress_votes[df_congress_votes.party == '더불어민주당']
df_kuk = df_congress_votes[df_congress_votes.party == '국민의당']
df_jus = df_congress_votes[df_congress_votes.party == '정의당']
df_etc = df_congress_votes.drop(pd.concat([df_sae, df_min, df_kuk, df_jus]).index)
```

```python
plt.figure(figsize=[60,50])
plt.gca().axes.get_xaxis().set_visible(False)
plt.gca().axes.get_yaxis().set_visible(False)

color_blue = '#4EACC5'
color_red = '#FF9C34' 
colors = map(lambda l:color_blue if l==0 else color_red, df_congress_votes['labels'].values)
plt.scatter(x=datapoints[:,0], y=datapoints[:,1], s=400,c=colors)
plt.scatter(x=centroids[:,0], y=centroids[:,1], s=400, marker='*', c=[color_blue, color_red])

blue_handle = mpatches.Patch(color=color_blue, label='cluster-0')
red_handle = mpatches.Patch(color=color_red, label='cluster-1')
sae_handle = mpatches.Patch(color='red', label='새누리당')
min_handle = mpatches.Patch(color='dodgerblue', label='더불어민주당')
kuk_handle = mpatches.Patch(color='lightgreen', label='국민의당')
jus_handle = mpatches.Patch(color='gold', label='정의당')
etc_handle = mpatches.Patch(color='silver', label='무소속')
plt.legend(handles=[blue_handle,red_handle,sae_handle,min_handle,kuk_handle,jus_handle,etc_handle])

def tag_name(name, xy, color):
    plt.annotate(name, 
        xy=xy, xycoords='data',
        xytext=xy+(-1,0), textcoords='data',
        size=15,
        alpha=0.5,
        ha='right', va='center',
        bbox={'boxstyle':'round,pad=0.5', 'fc':color},
        arrowprops={'relpos':(.5,.5), 'fc':'w', 'arrowstyle':'->', 'connectionstyle':'angle3'}
    )
    

for (n, xy) in zip(df_sae.name_kr.values , pca.transform(df_sae.iloc[:,3:-1])):
    tag_name(n, xy, 'red')
    
for (n, xy) in zip(df_min.name_kr.values, pca.transform(df_min.iloc[:,3:-1])):
    tag_name(n, xy, 'dodgerblue')
    
for (n, xy) in zip(df_kuk.name_kr.values, pca.transform(df_kuk.iloc[:,3:-1])):
    tag_name(n, xy, 'lightgreen')
    
for (n, xy) in zip(df_jus.name_kr.values, pca.transform(df_jus.iloc[:,3:-1])):
    tag_name(n, xy, 'gold')
    
for (n, xy) in zip(df_etc.name_kr.values, pca.transform(df_etc.iloc[:,3:-1])):
    tag_name(n, xy, 'silver')
```

## 20대 국회 지역구 당선자 표시

19대 국회 의결성향 클러스터링 결과에 20대 국회에 재선된 당선자들을 표시해봅니다. winners_tuple.pkl 은 20대 총선 직후에 언론에 발표된 지역구 의원들의 명단을 (당적, 이름) 튜플의 리스트로 가공해서 pickling 해둔 파일 입니다. 국회의원의 의결 행태가 국민들의 민의를 반영하는 직접선거를 통해 다시 당선되는 데 연관성이 있는지가 궁금했기에 이와 관계 없는 비례대표 당선자들은 반영하지 않았습니다. 19대 비례의원들 중에는 지역구로 공천받아 20대에 당선 되었거나 낙선된 경우에는 반영되어있는 샘입니다.

```python
winners_tup = pd.read_pickle('data/winners_tuple.pkl')
```

```python
winners_tup = set(winners_tup)
```

```python
pprint(winners_tup)
```

```python
def is_winner(row):
    return (row.party[0], row.name_kr) in winners_tup

df_congress_votes['winner'] = df_congress_votes.apply(is_winner, axis=1)

df_congress_votes['winner'].value_counts()
```

```python
df_sae = df_congress_votes[df_congress_votes.party == '새누리당']
df_min = df_congress_votes[df_congress_votes.party == '더불어민주당']
df_kuk = df_congress_votes[df_congress_votes.party == '국민의당']
df_jus = df_congress_votes[df_congress_votes.party == '정의당']
df_etc = df_congress_votes.drop(pd.concat([df_sae, df_min, df_kuk, df_jus]).index)
```

```python
plt.figure(figsize=[60,50])
plt.gca().axes.get_xaxis().set_visible(False)
plt.gca().axes.get_yaxis().set_visible(False)

color_blue = '#4EACC5'
color_red = '#FF9C34' 
colors = map(lambda l:color_blue if l==0 else color_red, df_congress_votes['labels'].values)
plt.scatter(x=datapoints[:,0], y=datapoints[:,1], s=400,c=colors)
plt.scatter(x=centroids[:,0], y=centroids[:,1], s=400, marker='*', c=[color_blue, color_red])

blue_handle = mpatches.Patch(color=color_blue, label='cluster-0')
red_handle = mpatches.Patch(color=color_red, label='cluster-1')
sae_handle = mpatches.Patch(color='red', label='새누리당')
min_handle = mpatches.Patch(color='dodgerblue', label='더불어민주당')
kuk_handle = mpatches.Patch(color='lightgreen', label='국민의당')
jus_handle = mpatches.Patch(color='gold', label='정의당')
etc_handle = mpatches.Patch(color='silver', label='무소속')
plt.legend(handles=[blue_handle,red_handle,sae_handle,min_handle,kuk_handle,jus_handle,etc_handle])

def tag_name(name, xy, color, win):
    alpha = 0.2 if not win else 1.0
    plt.annotate(name, 
        xy=xy, xycoords='data',
        xytext=xy+(-1,0), textcoords='data',
        size=15,
        ha='right', va='center',
        bbox={'boxstyle':'round,pad=0.5', 'fc':color, 'alpha':alpha},
        arrowprops={'relpos':(.5,.5), 'fc':'w', 'arrowstyle':'->', 'connectionstyle':'angle3'}
    )
    
for (n, xy, win) in zip(df_sae.name_kr.values, pca.transform(df_sae.iloc[:,3:-2]), df_sae.loc[:,'winner']):
    tag_name(n, xy, 'red', win)
    
for (n, xy, win) in zip(df_min.name_kr.values, pca.transform(df_min.iloc[:,3:-2]), df_min.loc[:,'winner']):
    tag_name(n, xy, 'dodgerblue', win)
    
for (n, xy, win) in zip(df_kuk.name_kr.values, pca.transform(df_kuk.iloc[:,3:-2]), df_kuk.loc[:,'winner']):
    tag_name(n, xy, 'lightgreen', win)
    
for (n, xy, win) in zip(df_jus.name_kr.values, pca.transform(df_jus.iloc[:,3:-2]), df_jus.loc[:,'winner']):
    tag_name(n, xy, 'gold', win)
    
for (n, xy, win) in zip(df_etc.name_kr.values, pca.transform(df_etc.iloc[:,3:-2]), df_etc.loc[:,'winner']):
    tag_name(n, xy, 'silver', win)

```
